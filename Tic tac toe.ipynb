```python
def show_board(board):
    print("".join(board[:3]))
    print("".join(board[3:6]))
    print("".join(board[6:]))

def verification(decision, board):
    return (decision.isdigit() 
            and 1 <= int(decision) <= 9 
            and board[int(decision)-1] == "-")

def check_if_win(board):
    options = [
        board[::3],
        board[1::3],
        board[2::3],
        board[:3],
        board[3:6],
        board[6:9],
        board[::4],
        board[2:7:2]
    ]
    return ["x"] * 3 in options or ["o"] * 3 in options
    

print("Tic tac toe!")

board = ["-"] * 9
player = "o"

# Main loop
while True:
    # Show board
    show_board(board)

    # Get decision and verify 
    decision = input(f"Where to put \'{[player]}\': ")
    if not verification(decision, board):
        print("Wrong entry - please insert a proper number (from 1 to 9)!")
        continue

    # Update the board
    board[int(decision)-1] = player

    # Verify for a win
    if check_if_win(board):
        # Show board with a win
        # As I check every round, latest player wins
        show_board(board)
        print(f"It's a win of \'{player}\'!")
        break

    # Verify if all fields populated
    if "-" not in board:
        # Show board with a draw
        show_board(board)
        print("Draw!")
        break

    # Switch players
    if player == "x":
        player = "o"
    else:
        player = "x"
```
